#### TENTANG ANALISIS RFM 
##### apa itu RFM?
RFM sendiri merupakan sebuah singkatan dari recency, frequency, dan monetary value. Recency menunjukkan waktu interaksi terakhir pelanggan dengan produk. Frequency adalah tentang seberapa sering pelanggan melakukan pembelian. Sedangkan, monetary value mengukur jumlah uang yang dihabiskan pelanggan untuk tiap transaksi pembelian.

Model analisis ini pertama kali diperkenalkan oleh Jan Roelf Bult dan Tom Wansbeek pada 1995 dalam sebuah artikel berjudul “Optimal Selection for Direct Mail” yang diterbitkan di jurnal bisnis Marketing Science. RFM analysis sering kali digunakan untuk membuktikan Prinsip Pemasaran Pareto atau Prinsip 80/20, yaitu, “20% jumlah pelanggan menghasilkan 80% pendapatan usaha”.

##### bagaimana cara kerjanya?
RFM analysis dapat dilakukan dengan dua cara: manual dan otomatis. Cara manual terkonsentrasi pada data demografis dan psikografis untuk bisa melakukan segmentasi pelanggan. Sayangnya, metode ini hanya mengandalkan data dari sampel dalam jumlah kecil.
Sementara itu, cara otomatis memanfaatkan tool khusus sehingga sampel bisa diambil secara otomatis. Salah satunya adalah dengan memanfaatkan sistem customer data platform (CDP). Dengan sistem tersebut, Anda bisa mengumpulkan dan menyatukan data mengenai perilaku pelanggan. Data yang diperoleh pun lebih terperinci sehingga saat Anda melakukan segmentasi pelanggan, hasil yang didapat lebih akurat.

Berikut adalah contoh pemanfaatan CRM (customer relationship management) tool dalam melakukan RFM analysis secara otomatis:

###### RECENCY
Seperti yang sudah Anda ketahui, recency menunjukkan waktu interaksi terakhir pelanggan dengan produk. Dalam metode otomatis, recency bisa menganalisis data mengenai transaksi terakhir pelanggan, kunjungan ke website produk, penggunaan aplikasi mobile, hingga jumlah “like” yang diberikan ke produk lewat media sosial

###### FREQUENCY
Frequency mengacu pada berapa kali pelanggan berinteraksi dengan produk atau melakukan transaksi dalam periode waktu tertentu. Frekuensi dijadikan ukuran karena dapat menunjukkan tingkat keterlibatan pelanggan dengan produk Anda. Pada tingkatan yang lebih tinggi, frekuensi juga bisa mengukur loyalitas pelanggan

###### MONETARY VALUE
Monetary value menunjukkan jumlah total yang dihabiskan oleh pelanggan untuk membeli produk Anda dalam periode waktu tertentu. Ia dianggap penting karena ada kecenderungan bahwa pelanggan yang sudah membelanjakan uangnya dalam jumlah besar akan membelanjakan lebih banyak lagi di masa mendatang


#### kelebihan RFM
RFM analysis akan membantu pemilik bisnis meningkatkan pendapatan dengan menargetkan kelompok tertentu dari pelanggan mereka. Dengan menerapkan analisis ini, maka segmentasi pelanggan akan lebih terpersonalisasi, menyesuaikan perilaku pelanggan pada transaksi di waktu sebelumnya.

Misalnya, strategi bundling produk akan lebih efektif jika ditawarkan pada pelanggan yang sering berbelanja dalam jumlah besar. Di sisi lain, strategi membership lebih cocok diterapkan pada pelanggan yang sering mengakses akun media sosial produk.

Sejak pertama kali diperkenalkan hingga sekarang, RFM analysis telah terbukti mampu memprediksi perilaku pelanggan di masa mendatang secara efektif. Jika hal tersebut kemudian dikelola dengan baik, bukan tidak mungkin pendapatan bisnis akan meningkat...

#### KMEANS++
K-Means adalah sebuah algoritma pengelompokan data yang digunakan untuk melakukan segmentasi data menjadi kelompok-kelompok yang serupa berdasarkan karakteristik atau atribut yang dimiliki oleh data tersebut. Algoritma ini adalah salah satu metode clustering yang paling populer dan sederhana yang digunakan dalam analisis data.

Idea dasar dari K-Means adalah membagi data menjadi K kelompok atau kluster, di mana setiap kluster diwakili oleh sebuah centroid (pusat kelompok). Setiap titik data akan dikelompokkan ke kluster yang memiliki centroid terdekat. Proses ini dilakukan berulang hingga centroid tidak berubah atau mencapai kriteria penghentian yang ditentukan.

Mengapa K-Means digunakan untuk segmentasi pelanggan? Berikut adalah beberapa alasan:

- Tidak memerlukan label: K-Means adalah metode pengelompokan yang tanpa pengawasan (unsupervised), artinya tidak memerlukan label atau informasi sebelumnya tentang kelompok mana yang sudah ada. Dalam segmentasi pelanggan, kita seringkali tidak memiliki informasi sebelumnya tentang kelompok-kelompok yang ada, sehingga K-Means cocok digunakan.

- Skalabilitas: K-Means memiliki kinerja yang baik dan skalabilitas yang tinggi. Ini berarti algoritma ini dapat diterapkan pada dataset yang cukup besar dengan efisiensi yang baik, sehingga cocok digunakan untuk segmentasi pelanggan di perusahaan dengan jumlah pelanggan yang besar.

- Interpretasi yang sederhana: Hasil K-Means cukup mudah untuk diinterpretasikan. Setiap kluster memiliki centroid yang mewakili karakteristik kelompok tersebut. Dengan menganalisis dan memahami atribut-atribut yang membedakan kluster-keluster tersebut, kita dapat memperoleh wawasan yang berharga t- entang perilaku atau preferensi pelanggan.

Pengambilan keputusan: Segmentasi pelanggan dengan menggunakan K-Means dapat membantu perusahaan dalam pengambilan keputusan strategis, seperti penargetan pasar, personalisasi layanan, atau pengembangan strategi pemasaran yang lebih efektif.

### OUTLIER DATA
metode yang di gunakan mendeteksi outlier adalah metode IQR. metode IQR memenafaatkan quantil 25% dan quantil 75%. 

```python
df.describe()
```

```python
df.info()
```

quantile 25% dan 75% dapat di manfaatkan sebagai acuan untuk mencari outlier menggunakan IQR

```python
##mencari IQR(outlier) harga

Q1 = df['harga'].quantile(0.25)
Q3 = df['harga'].quantile(0.75)
IQR = Q3 - Q1

outliers = df[(df['harga'] < (Q1 - 1.5 * IQR)) | (df['harga'] > (Q3 + 1.5 * IQR))]
print("jumlah outlier di column harga sebanyak:", len(outliers))
```

```python
#menghapus outlier

df = df[~((df['harga'] < (Q1 - 1.5 * IQR)) | (df['harga'] > (Q3 + 1.5 * IQR)))]
```

```python
#melakukan hal yang sama pada column jumlah barang yang terjual

Q1j = df['jumlah'].quantile(0.25)
Q3j = df['jumlah'].quantile(0.75)
IQRj = Q3j - Q1j

outliers2 = df[(df['jumlah'] < (Q1j - 1.5 * IQRj)) | (df['jumlah'] > (Q3j + 1.5 * IQRj))]
print("jumlah outlier di column jumlah(quantity) sebanyak:", len(outliers2))
```

```python
##menghapus outlier

df = df[~((df['jumlah'] < (Q1j - 1.5 * IQRj)) | (df['jumlah'] > (Q3j + 1.5 * IQRj)))]
```

```python
#membuat features baru yaitu total pendapatan yang merupakan hasil perkalian dari jumllah barang yang di beli dan harga

df['total'] = df['jumlah'] * df['harga']
```

```python
#melakukan pengelompokan data berdasarkan tingkat revenue dan akan di beri nama revenue lavel

def revenuelevel(row) :
    if row ['total'] <= 10:
      return 'low revenue'
    elif row['total'] >10 and row['total'] <30:
      return 'middle revenue'
    else:
      return 'high revenue'


df['revenuelevel'] = df.apply(revenuelevel, axis = 1)
```

#### MEMBUAT LINE CHART DENGAN ESKSPONENSIAL SMOATHING
Exponential Smoothing adalah teknik smoothing yang memberikan bobot yang berkurang 
secara eksponensial kepada titik data sebelumnya

```python
alpha= 0.1
eksponensial = df.groupby(['timeline'])['total'].sum()
ses = pd.Series(eksponensial).ewm(alpha=alpha).mean()
```

```python
plt.figure(figsize=(14,7))
plt.plot(eksponensial, label='Data Asli', color='teal')
plt.plot(ses, label=f'Exponential Smoothing (α={alpha})', color='crimson', linewidth=3)

# Customize plot appearance
plt.legend()
plt.title('TRANSACTION TIMELINE(Exponential Smoothing)')
plt.xlabel('Waktu')
plt.ylabel('pendapatan penjualan')
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)


plt.savefig("timeline_transaksi1EXPONENTIAL.png", dpi=300)
plt.show()

```

## RFM MODEL

mengasumsikan tanggal sekarang merupakan 1 hari setelah transaksi terakhir terjadi agar angka resensi tidak terlalu besar.

```python
sekarang = pd.to_datetime('2011-12-10')
```

```python
df.sample()
```

```python
##membuat model RFM

table = df.groupby('idcustomer').agg({'timeline': lambda x: (sekarang - x.max()).days,
                                      'idcustomer': lambda x: len(x),
                                      'total': lambda x: x.sum()})
```

```python
#mengubah tipe data

table['timeline'] = table['timeline'].astype(int)
```

```python
#mengganti nama variable

table.rename(columns= {'timeline': 'recency', 'idcustomer': 'frequency', 'total': 'monetary'}, inplace= True)
```

```python
 #melihat table yang sudah di buat

table.sample(5)
```

tabel diatas merupakan tabel RFM yang akan di gunakan untuk analisis

```python
##melakukan segmentasi pelanggan dengan memanfaatkan quantile pada data

segmentasi = table.quantile(q=[0.25,0.5,0.75])
```

```python
segmentasi
```

tabel di atas merupakan acuan dasar untuk melakukan segmentasi terhadap customer

```python
segmentasi = segmentasi.to_dict()
```

```python
segmentasirfm = table
```

```python
# menggunan fungsi def untuk segmentasi

def rclass(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]:
        return 3
    else:
        return 4
```

```python
def fmclass(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]:
        return 2
    else:
        return 1
```

```python
segmentasirfm['R_quartile'] = segmentasirfm['recency'].apply(rclass, args=('recency',segmentasi,))
segmentasirfm['F_quartile'] = segmentasirfm['frequency'].apply(fmclass, args=('frequency',segmentasi,))
segmentasirfm['M_quartile'] = segmentasirfm['monetary'].apply(fmclass, args=('monetary',segmentasi,))
```

```python
segmentasirfm['RFMclass'] = segmentasirfm.R_quartile.map(str) \
                            + segmentasirfm.F_quartile.map(str) \
                            + segmentasirfm.M_quartile.map(str)
```

```python
# melakukan cek ulang fungsi def yang sudah di buat

segmentasirfm.sample(10)
```

```python
segmentasirfm['RFMclass'] = segmentasirfm['RFMclass'].astype('int')
segmentasirfm['recency'] = segmentasirfm['recency'].astype('int')
```

#### melakukan grouping RFM dan customer type

```python
def RFM_type(row) :
    if row ['RFMclass'] <=150:
        return 'excellent RFM'
    elif row ['RFMclass'] >150 and row ['RFMclass'] <= 250:
        return 'good RFM'
    elif row ['RFMclass'] >250 and row ['RFMclass'] <= 360:
        return 'moderate RFM'
    else :
        return 'Bad RFM'
```

```python
segmentasirfm['RFM_type'] = segmentasirfm.apply(RFM_type, axis=1)  
```

melakukan segmentasi pelanggan dengan mengkategorikan customer berdasarkan nilai rfm yang di sudah di lakukan sebelumnya sehingga di dapatkan kategori customer sebagai berikut:
- customer dengan nilai RFMclass di bawah 150 akan di nyatakan sebagai excellent RFM.
- customer dengan nilai RFMclass di antara 151 sampai 250 akan di nyatakan sebagai good RFM.
- customer dengan nilai RFMclass di antara 250 sampai 360 akan dinyatakam sebagai moderate RFM.
- custoemr dengan nilai RFMclass di atas 360 akan dinyatakan sebagai bad RFM.

```python
def customer_type(row) :
    if row ['monetary'] <=1000:
        return 'normal'
    elif row ['monetary'] >1000 and row ['monetary'] <= 2000:
        return 'silver'
    elif row ['monetary'] >2000 and row ['monetary'] <= 3000:
        return 'gold'
    else :
        return 'diamond'

segmentasirfm['customer_class'] = segmentasirfm.apply(customer_type, axis=1)
```

melakukan segmentasi pelanggan dengan mengkategorikan customer berdasarkan uang yang sudah di spent atau di keluarkan oleh customer, berikut adalah kategori customer yang di lakukan:
- customer dengan nilai monetary di bawah 1000 akan di nyatakan sebagai normal customers.
- customer dengan nilai monetary di antara 1000 sampai 2000 akan di nyatakan sebagai silver customers.
- customer dengan nilai monetary di antara 2000 sampai 3000 akan dinyatakam sebagai gold customers.
- custoemr dengan nilai monetary di atas 3000 akan dinyatakan sebagai diamond customers.

## K- MEANS CLUSTERING
K-means clustering adalah algoritma pembelajaran mesin yang populer digunakan untuk mengelompokkan atau mengklasterkan titik-titik data yang serupa. Tujuan dari algoritma ini adalah untuk membagi dataset yang diberikan menjadi K klaster, di mana K adalah jumlah yang telah ditentukan sebelumnya oleh pengguna.

Algoritma ini bekerja dengan cara secara iteratif menetapkan titik-titik data ke dalam klaster dan memperbarui pusat klaster hingga konvergennsi
i.

```python
datamodel = segmentasirfm.drop({'R_quartile', 'F_quartile',	'M_quartile',	'RFMclass',	'RFM_type',	'customer_class'}, axis=1)
```

```python
datamodel =  datamodel.drop('unknown', axis=0)
```

penghapusan outlier data

```python
sns.boxplot(data=datamodel, y='frequency')
```

```python
box2 = datamodel.select_dtypes('int', 'float')
```

```python
datamodel = datamodel[datamodel['frequency'] < 190].reset_index(drop = True)
```

```python
datamodel.count()
```

```python
#melakukan standarisasi data

scaler = StandardScaler()
datamodel_scaled = scaler.fit_transform(datamodel)
```

```python
#membuat fungsi elbow method

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(datamodel_scaled)
    wcss.append(kmeans.inertia_)


#melakukan visualisasi elbow method

sns.set(style='whitegrid')

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(range(1, 11), wcss, marker='o', linestyle='-', color='blue')
axes[0].set_title('Elbow Method for Determining Number of Clusters')
axes[0].set_xlabel('Jumlah Cluster')
axes[0].set_ylabel('WCSS')

axes[1].plot(range(1, 11), wcss, marker='o', linestyle='-', color='blue')
axes[1].scatter(3, wcss[2], marker='o', color='red', label='Elbow Point')
axes[1].annotate('Elbow Point', xy=(3, wcss[2]), xytext=(3.5, wcss[2]+200), 
                 arrowprops=dict(facecolor='black', arrowstyle='->'))
axes[1].set_title('Elbow Method with Elbow Point')
axes[1].set_xlabel('Jumlah Cluster')
axes[1].set_ylabel('WCSS')
axes[1].legend()

plt.tight_layout()
plt.savefig("elbowmethod", dpi=300)
plt.show()
```

jumlah cluster terbaik untuk data rfm models yang akan di gunakan di algortima KMeans menurut elbow methods berjumlah 4 cluster

```python
## meneraokan model clastering KMeans++
kmeans= KMeans(n_clusters=3, init='k-means++', max_iter=1000, n_init=20,
              random_state=0)
cluster_labels = kmeans.fit_predict(datamodel_scaled)
```

```python

# Mendefinisikan label kluster
cluster_labels = kmeans.fit_predict(datamodel_scaled)
cluster_centers = kmeans.cluster_centers_

# Plotting scatter plot untuk data
plt.figure(figsize=(10, 10))
scatter = plt.scatter(datamodel_scaled[:, 0], datamodel_scaled[:, 1], c=cluster_labels, cmap='Set1', alpha=0.8)

# Plotting scatter plot untuk centroid
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='black', marker='X', s=100, label='Centroid')

plt.xlabel('Fitur 1')
plt.ylabel('Fitur 2')
plt.title('Visualization of K-means Clustering Results', fontsize=14)

plt.xlim([datamodel_scaled[:, 0].min() - 0.5, datamodel_scaled[:, 0].max() + 0.5])
plt.ylim([datamodel_scaled[:, 1].min() - 0.5, datamodel_scaled[:, 1].max() + 0.5])

# Membuat legenda untuk masing-masing warna kluster
legend_labels = []
for label in set(cluster_labels):
    legend_labels.append(f'Kluster {label}')

plt.legend(handles=scatter.legend_elements()[0], labels=legend_labels)

plt.savefig("kmeans++clustering.png", dpi=300)
plt.show()
```

untuk data yang sedang di analisis, bentuk cluster yang di olah dengan algoritma kmeans kurang baik karena tidak memiliki bentuk bola namun pada kasus ini kita akan lanjut untuk sekedar eksplorasi saja
